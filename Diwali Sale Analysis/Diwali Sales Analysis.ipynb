```python
# Number of customers by gender - 

sns.countplot(x='Gender', data=df, hue ='Gender', palette=['Red', 'Blue'])
plt.title("Count Of Customers By Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Purchasing power by Gender

sales_gender = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gender, hue ='Gender', palette=['Red', 'Blue'])
plt.title("Gender-wise Purchasing Power")
plt.xlabel("Gender")
plt.ylabel("Amount Spent")
plt.show()
```

```python
# Based on Age -
```

```python
# Number of customers by Age Group - 

sns.countplot(x='Age Group', data=df, hue ='Gender', palette=['Red', 'Blue'])
plt.title("Count Of Customers By Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.ylabel("Count")
plt.show()
```

```python
# Purchasing power by Age Group

sales_gender = df.groupby(["Age Group", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x='Age Group', y='Amount', hue='Gender', data=sales_gender, palette=['Red', 'Blue'])
plt.title("Purchasing Power by Age Group and Gender")
plt.xlabel("Age Group")
plt.ylabel("Amount Spent")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.legend(title="Gender")
plt.show()
```

```python
# By Occupation - 
```

```python
# Occupation-wise spending habit -

sales_Occupation = df.groupby(["Occupation", "Product_Category"])["Amount"].sum().reset_index()

pivot_table = sales_Occupation.pivot(index="Occupation", columns="Product_Category", values="Amount")

# Setting up the plot
sns.set_style("dark")
plt.figure(figsize=(15, 10))  # Adjusted figure size for better visibility

# Create the heatmap
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlOrRd", cbar_kws={'label': 'Total Amount'})

plt.title("Total Sales Amount by Occupation and Product_Category")
plt.xlabel("Product_Category")
plt.ylabel("Occupation")

# Rotate x-axis labels 
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# By State - 
```

```python
# Pie chart showing the state-wise contribution to total sales - 

sales_states = df.groupby(['State'])['Amount'].sum()


# Setting up the plot
plt.figure(figsize=(10, 8))  # Adjusted figure size for better visibility



# Adjust label distance:
plt.pie(sales_states, labels=sales_states.index, autopct="%1.1f%%",labeldistance=1.1,textprops={'fontsize': 5})



plt.title("Sales Per State")  # Add a title for clarity
plt.show()
```

```python
# Pie chart showing the category-wise contribution to total sales - 

sales_product_category = df.groupby(['Product_Category'])['Amount'].sum()


# Setting up the plot
plt.figure(figsize=(10, 8))  # Adjusted figure size for better visibility

# Plot the pie chart with adjustments
plt.pie(sales_product_category, labels=sales_product_category.index, autopct="%1.1f%%", textprops={'fontsize': 5})

plt.title("Sales Per Product Category")  # Add a title for clarity
plt.show()
```

```python
# Top 10 customers by ordered amount - 

sales_product_category = df.groupby('Customer_name')['Amount'].sum()

# 2. Sort by amount spent (descending order for top spenders)
top_10_customers = sales_product_category.sort_values(ascending=False).head(10)  # Get top 10

# 3. Create the bar chart using seaborn
sns.barplot(x=top_10_customers.index, y=top_10_customers.values)  # Use top_10_customers for data

plt.title("Top 10 Customers (Amount Spent)")
plt.xlabel("Customer Name")
plt.ylabel("Amount Spent")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust spacing for better readability
plt.show()
```

```python
# Scatterplot showing the spending habit based in Age and Gender - 
sns.set_style("whitegrid")

# Create the scatter plot (no need for axes argument)
sns.scatterplot(data=df, x='Age', y='Amount', hue='Gender')

# Add title and labels
plt.title('Age vs Amount')
plt.xlabel('Age')
plt.ylabel('Amount')

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()

```
